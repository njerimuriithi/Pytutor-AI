```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()
```

```python
groq_client = Groq(api_key=os.getenv("API_KEY"))
```

```python
topics = ["Python Functions", "OOP", "Loops"]

prompt = f"""
Start an interactive Python quiz.
Topics: {", ".join(topics)}

Rules:
- Ask one multiple choice question at a time.
- Provide 4 options (A, B, C, D).
- Wait for the student's answer.
- Confirm if correct or incorrect and explain.
- Then ask the next question.
"""
```

```python
messages = [
    {
        "role": "system",
        "content": """
You are a Python tutor that gives interactive quizzes.

Rules:
- Ask one multiple choice question at a time.
- Wait for the student's answer.
- If correct, say "Correct!" and explain.
- If wrong, say "Incorrect" and explain the correct answer.
- Then ask the next question.
"""
    },
    {
        "role": "user",
        "content": prompt
    }
]

while True:
    completion = groq_client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=messages,
        temperature=0.3,
        max_completion_tokens=8192,
        top_p=1,
        reasoning_effort="medium",
        stream=True,
        stop=None
    )

    assistant_reply = ""

    for chunk in completion:
        content = chunk.choices[0].delta.content or ""
        print(content, end="")
        assistant_reply += content

    print("\n")

    # Save assistant response to history
    messages.append({
        "role": "assistant",
        "content": assistant_reply
    })

    # Get student answer
    user_input = input("Your answer: ")

    # Add student answer to conversation
    messages.append({
        "role": "user",
        "content": user_input
    })
```
